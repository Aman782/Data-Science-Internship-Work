```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
y = df['DEATH_EVENT']
x = df.drop(columns='DEATH_EVENT')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=88)
```

```python
svm_model = svm.SVC()
```

```python
svm_model.fit(x_train, y_train)
```

```python
print("Train accuracy:", svm_model.score(x_train, y_train))
print("Test accuracy:", svm_model.score(x_test, y_test))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc_fit = sc.fit(x_train)
```

```python
x_train_sc = sc_fit.transform(x_train)
```

```python
sc_fit_test = sc.fit(x_test)
```

```python
x_test_sc = sc_fit_test.transform(x_test)
```

```python
x_train_sc
```

```python
x_train1 =  pd.DataFrame(x_train_sc, columns=x_train.columns)
```

```python
x_train1
```

```python
x_test1 =  pd.DataFrame(x_test_sc, columns=x_test.columns)
```

```python
x_test1
```

```python
svm_model.fit(x_train1, y_train)
```

```python
print("Training accuracy:", svm_model.score(x_train1, y_train))
print("Testing accuracy:", svm_model.score(x_test1, y_test))
```

```python
svm_model = svm.SVC(C=1.0, kernel='linear', degree=3)
```

```python
svm_model.fit(x_train1, y_train)
```

```python
print("Training accuracy:", svm_model.score(x_train1, y_train))
print("Testing accuracy:", svm_model.score(x_test1, y_test))
```
