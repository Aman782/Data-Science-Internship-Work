# matplotlib 

```python
import matplotlib.pyplot as plt
```

```python
l = [1, 4, 2, 7, 5]
l2 = [i**2 for i in l]
l3 = [i**3 for i in l]
```

```python
# plt.plot("x-axis", "y-axis")
```

# Line Chart

```python
plt.plot(l)
```

```python
plt.plot(l, "*") # "*" for marker inplace of line
```

```python
plt.plot(l, "o") # "o" for marker inplace of line
```

```python
plt.plot(l, "o", markersize=20) # markersize for increasing the markersize
```

plt.plot(l, "o--", color = "black")

# we can plot multiple chart on a single plot
plt.plot(l)
plt.plot(l2)
plt.plot(l3)

# Line Chart for 2 Data points

```python
plt.plot(l, l2, "o--")
```

```python
plt.plot(l2, l3, "o--")
plt.xlabel("l2_on_X-axis")
plt.ylabel("l3_on_Y-axis")
plt.show()
```

```python
plt.figure(figsize=(10, 4)) # To increase the size of figure
plt.plot(l, l3)
plt.show()
```

```python
plt.plot(l)
plt.plot(l3)
```

```python
plt.plot(l3, "o--", markersize=15, color="k")
plt.plot(l2, "*--", markersize=20, color="b")
plt.show()
```

```python
plt.plot(l2, label = "l2")
plt.plot(l3, label = "l3")
plt.legend()
plt.show()
```

# Bar Chart

```python
st = ["a", "b", "c", "d"]
m = [22, 60, 45, 30]
plt.xlabel("Students_names")
plt.ylabel("marks")
plt.bar(st, m, color=["k", "b", "r"])
plt.show()
```

# Adding data label on the bars

```python
st = ["a", "b", "c", "d"]
m = [22, 60, 45, 30]
plt.xlabel("Students_names")
plt.ylabel("marks")
newbar = plt.bar(st, m, color=["k", "b", "r"])
plt.bar_label(newbar)
plt.plot(l2, "o--", markersize=20, color="yellow")
plt.show()
```

# Pie Chart 

```python
# Data
sizes = [20, 35, 15, 30]
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']

# Create pie chart
plt.pie(sizes, labels=labels)
```

```python
plt.pie(sizes, labels=labels, explode=[.2, 0, 0, 0])
```

# Plot a bar chart having 5 values and add a line with different color marker and different color line do mention the xlabel, 
# title and ylabel also

```python
persons = ["a", "b", "c", "d", "e"]
height = [159, 210, 137, 135, 110]
age = [22, 34, 55, 27, 30]
plt.xlabel("persons")
plt.ylabel("height(in cms)")
b = plt.bar(persons, height, color=["yellow", "r", "k", "b", "g"])
plt.bar_label(b)
plt.plot(age, "o--", markersize=10,color="cyan")
```
