# Removing the outliers from all numerical data

```python
def calcu_lf_uf(q1, q3):
   IQR = q3-q1
   lower_fence = q1-(1.5*IQR)
   upper_fence = q3+(1.5*IQR)

   print("if-->", lower_fence, "uf-->", upper_fence)    
   return [lower_fence, upper_fence]
```

```python
df['Units_Sold'].describe()
```

```python
[lf, uf] = calcu_lf_uf(2456.250, 7574.500)
```

```python
df['Units_Sold'] = np.where(df['Units_Sold']>uf, uf, df['Units_Sold'])
df['Units_Sold'] = np.where(df['Units_Sold']<lf, lf, df['Units_Sold'])
```

```python
plt.boxplot(df['Units_Sold'])
plt.show()
```
