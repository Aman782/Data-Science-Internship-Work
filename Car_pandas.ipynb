```python
df["Sales_in_thousands"].describe()
```

```python
q1 = 14.114
q3 = 67.956
IQR = q3-q1

lower_fence = q1-(1.5*IQR)
upper_fence = q3+(1.5*IQR)

print("if-->", lower_fence, "uf-->", upper_fence)
```

```python
df["Sales_in_thousands"] = np.where(df["Sales_in_thousands"]> upper_fence, upper_fence, df["Sales_in_thousands"])
```

```python
plt.boxplot(df["Sales_in_thousands"])
plt.show()
```
