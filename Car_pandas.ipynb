# Removing the outlier from numerical column

```python
# Sales_in_thousands
# four_year_resale_value
# Price_in_thousands
# Engine_size	
# Horsepower
# Wheelbase
# Width
# Length
# Curb_weight 
# Fuel_capacity
# Fuel_efficiency
# Power_perf_factor
```

```python
# Removing the outliers from 'Sales in thousands' column
def calcu_lf_uf(q1, q3):
   IQR = q3-q1
   lower_fence = q1-(1.5*IQR)
   upper_fence = q3+(1.5*IQR)

   print("if-->", lower_fence, "uf-->", upper_fence)
   return [lower_fence, upper_fence]
```

```python
calcu_lf_uf(14.114, 67.956)
```

```python
df["Sales_in_thousands"] = np.where(df["Sales_in_thousands"]> upper_fence, upper_fence, df["Sales_in_thousands"])
```

```python
plt.boxplot(df["Sales_in_thousands"])
plt.show()
```
